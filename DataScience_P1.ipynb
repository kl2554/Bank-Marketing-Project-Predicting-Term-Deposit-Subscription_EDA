**Visualize the Data**
1.Let's create some visualizations to better understand the distributions and relationships in the data.

4.1. Distribution of Numerical Features

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

for feature in numerical_features:
    plt.figure(figsize=(2, 1))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

4.3. Categorical Feature Analysis
Analyze categorical features using bar plots

```python
# Plot count of categorical features
categorical_features = df.select_dtypes(include=['object']).columns

for feature in categorical_features:
    plt.figure(figsize=(2, 1))
    sns.countplot(data=df, x=feature)
    plt.title(f'Count of {feature}')
    plt.xticks(rotation=45)
    plt.show()

```
